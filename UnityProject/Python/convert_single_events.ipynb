```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("../Assets/Data/202401-citibike-tripdata_1.csv")
data = data.dropna(axis= 0)
print(data.describe())

start_data = data.filter(["ride_id", "rideable_type", "started_at","start_station_name","start_station_id","start_lat","start_lng","member_casual"],axis= 1)
start_data.rename(columns={"started_at": "time", "start_station_name": "station_name", "start_station_id" :"station_id",
                "start_lat": "lat", "start_lng" : "lng"},inplace=True)
start_data["state"] = 0
end_data = data.filter(["ride_id", "rideable_type", "ended_at","end_station_name","end_station_id","end_lat","end_lng","member_casual"],axis= 1)
end_data.rename(columns={"ended_at": "time", "end_station_name": "station_name", "end_station_id" :"station_id",
                "end_lat": "lat", "end_lng" : "lng"},inplace=True)
end_data["state"] = 1
data = pd.concat([start_data,end_data])

data["lat"] = (data["lat"] - np.min(data["lat"]))/(np.max(data["lat"]) - np.min(data["lat"]))
data["lng"] = (data["lng"] - np.min(data["lng"]))/(np.max(data["lng"]) - np.min(data["lng"]))

print(np.min(data["lat"]))
print(np.max(data["lat"]))
print(np.min(data["lng"]))
print(np.max(data["lng"]))
data["time"] = pd.to_datetime(data['time'])
data = data.sort_values(by='time')
data
```

```python
data = data.reindex(["ride_id","rideable_type","time", "station_name", "station_id", "member_casual","lat","lng", "state"],axis = 1)
data.to_csv("../Assets/Data/Event_Data.csv", index=False)
```
